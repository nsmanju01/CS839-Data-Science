```python
# Import packages
import os
import pandas as pd
import numpy as np
```

```python
#Obtain the path 
data_set_path = os.getcwd() + os.sep + "data"
```

```python
#Create path to read file
data_set_file_path = data_set_path + os.sep + "matched_data.csv"
merged_data_set_file_path = data_set_path + os.sep + "merged_data.csv"
```

```python
#Read CSV file into DataFrame
print data_set_file_path
matched = pd.read_table(data_set_file_path,sep=',')
```

```python
#Create empty DataFrame of merged schema 
columns = ['name','release_year', 'certificate','genre','director','actors','runtime','gross','production_house']
merged = pd.DataFrame(columns = columns)
print merged
```

```python
#Create column lists for creating new merged DataFrames
name = []
release_year = []
certificate = []
runtime = []
genre = []
director = []
gross = []
actors = []
production_house = []

actorsSet = set()
directorsSet = set()

#Loop over every row of the matched table to select the appropriate field for the merged columns
for index, row in matched.iterrows():
   
    #Name: Select longest length name
    name1 = row['name_x']
    name2 = row['name_y']
    if(len(name1)>len(name2)):
        name.append(name1)
    else:
        name.append(name2)
    
    #Release Year: Select latest year
    year1 = row['release_year_x']
    year2 = row['release_year_y']
    if(year1>year2):
        release_year.append(year1)
    else:
        release_year.append(year2)
        
    #Certificate: Select smallest length certificate 
    cert1 = row['certificate_x']
    cert2 = row['certificate_x']
    if(len(cert1)>len(cert2)):
        certificate.append(cert1)
    else:
        certificate.append(cert2)
        
    #Runtime: Select longest runtime
    run1 = row['runtime_x']
    run2 = row['runtime_y']
    if(run1>run2):
        if run1 == 'None':
            run1 = 0
        runtime.append(int(run1))
    else:
        if run2 == 'None':
            run2 = 0
        runtime.append(int(run2))
        
    #Genre: Select smaller length genre 
    gen1 = row['genre_x']
    gen2 = row['genre_y']
    if(len(gen1)<len(gen2)):
        if '_' in gen1:
            gen1 = gen1.split('_')[1]
        genre.append(gen1)
    else:
        if '_' in gen2:
            gen2 = gen2.split('_')[1]
        genre.append(gen2)
        
    #Director: Select longest length director
    dir1 = row['director_x']
    dir2 = row['director_y']
    if(len(dir1)>len(dir2)):
        director.append(dir1)
    else:
        director.append(dir2)
        
    #Gross: There is only one
    gross.append(row['gross_x'])
    
    #Actors: Select longest length actor
    act1 = row['actors_x']
    act2 = row['actors_y']
    if(len(act1)>len(act2)):
        actors.append(act1)
    else:
        actors.append(act2)
    
    #Production house: There is only one
    production_house.append(row['production_house'])
#print genre
print len(production_house)   

for actorRow in actors:    
    for actor in actorRow.split('_'):
        actorsSet.add(actor)
        
print actorsSet
print len(actorsSet)
```

```python
def returnCount(df,attr,flag):
    mydict = {}
    for index, row in df.iterrows():
        entry = row[attr]
        list = []
        if(flag==0):
            list = entry.split()
        else:
            list.append(entry)
        for word in list: 
            if(word not in mydict.keys()):
                mydict[word] = 1
            else:
                mydict[word] = mydict[word] + 1
    return mydict;

```

```python
#Now merge all the columns to the merged table
merged['name'] = name
merged['release_year'] = release_year
merged['certificate'] = certificate
merged['runtime'] = runtime
merged['genre'] = genre
merged['director'] = director
merged['gross'] = gross
merged['actors'] = actors
merged['production_house'] = production_house
print merged
```

```python
#Save the merged DataFrame to file as csv format
merged.to_csv(merged_data_set_file_path)
print "File saved as ",merged_data_set_file_path 
```

```python
from collections import Counter
def topTenDict(mydict):
    d = Counter(mydict)
    for k, v in d.most_common(10):
        print '%s: %i' % (k, v)

```

```python

#get actor wise gross count
print "hh"
actorGrossDict = dict()
for index, row in merged.iterrows():
    for actr in actorsSet:
        if actr in row['actors']:
            if actr in actorGrossDict.keys():
                actorGrossDict[actr] += row['gross']
            else:
                actorGrossDict[actr] = row['gross']
                
print "jay"          
#print actorGrossDict
topTenDict(actorGrossDict)
```
